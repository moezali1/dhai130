# Dataset

```python
from pycaret.datasets import get_data
data = get_data('diabetes')
```

**Context** </br>
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

**Content** </br>
The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

**Acknowledgement** </br>
Smith, J.W., Everhart, J.E., Dickson, W.C., Knowler, W.C., & Johannes, R.S. (1988). Using the ADAP learning algorithm to forecast the onset of diabetes mellitus. In Proceedings of the Symposium on Computer Applications and Medical Care (pp. 261--265). IEEE Computer Society Press.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is the pandas DataFrame containing your data

# Extracting the 'Carat Weight' and 'Price' columns
carat_weight = data['Carat Weight']
price = data['Price']

# Creating the scatter plot
plt.scatter(carat_weight, price)

# Adding labels and title
plt.xlabel('Carat Weight')
plt.ylabel('Price')
plt.title('Scatter Plot: Carat Weight vs Price')

# Displaying the plot
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Assuming 'data' is the pandas DataFrame containing your data

# Creating the scatter plot using Plotly
fig = px.scatter(data, x='Carat Weight', y='Price')

# Adding labels and title
fig.update_layout(
    xaxis_title='Carat Weight',
    yaxis_title='Price',
    title='Scatter Plot: Carat Weight vs Price'
)

# Displaying the plot
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Assuming 'data' is the pandas DataFrame containing your data

# Creating the scatter plot using Plotly
fig = px.scatter(data, x='Carat Weight', y='Price', color='Report',
                 hover_data=['Color', 'Clarity'], opacity=0.7,
                 title='Scatter Plot: Carat Weight vs Price',
                 labels={'Carat Weight': 'Carat Weight (ct)', 'Price': 'Price ($)'})

# Customizing marker size, style, and color
fig.update_traces(marker=dict(size=8, line=dict(width=0.5, color='White')),
                  selector=dict(mode='markers'))

# Customizing axis labels and title
fig.update_layout(xaxis=dict(tickformat='.2f'), yaxis=dict(tickformat=','),
                  xaxis_title='Carat Weight', yaxis_title='Price')

# Customizing legend position and title
fig.update_layout(legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
                  legend_title='Cut')

# Adding grid lines
fig.update_layout(xaxis=dict(showgrid=True, gridwidth=0.5, gridcolor='lightgray'),
                  yaxis=dict(showgrid=True, gridwidth=0.5, gridcolor='lightgray'))

# Adding a custom color scale
fig.update_layout(coloraxis_colorbar=dict(title='Cut', tickvals=[1, 2, 3, 4, 5],
                                          ticktext=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']))

# Displaying the plot
fig.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate average price for each cut category
average_price = data.groupby('Cut')['Price'].mean()

# Generate bar plot
plt.bar(average_price.index, average_price.values)
plt.xlabel('Cut')
plt.ylabel('Average Price')
plt.title('Average Price by Cut')
plt.show()
```

```python
import plotly.express as px

# Calculate average price for each cut category
average_price = data.groupby('Cut')['Price'].mean().reset_index()

# Generate bar plot using Plotly
fig = px.bar(average_price, x='Cut', y='Price', labels={'Cut': 'Cut', 'Price': 'Average Price'})
fig.update_layout(title='Average Price by Cut')

fig.show()
```
